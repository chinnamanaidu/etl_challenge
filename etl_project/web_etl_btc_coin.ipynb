### This notebook is to get the cryptocurrency data from yahoo website

```python
import pandas as pd
from sqlalchemy import create_engine
from splinter import Browser
from bs4 import BeautifulSoup as bs
import time
```

# Get and initialize the chrome driver to load the page

```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

### Initialize the Browser

```python
# Visit visitcostarica.herokuapp.com
url = "https://finance.yahoo.com/cryptocurrencies"
browser.visit(url)
### Initialize the Browser
time.sleep(1)

# Scrape page into Soup
```

### screen scrap the data, this section goes through the yahoo website to get the data, there are two level of iterations, the first level is to load each data page and second level is to get the required data sets

```python
html = browser.html
soup = bs(html, "html.parser")
data_array = []
avg_temps = soup.find("div", class_="Ovx(a) Ovx(h)--print Ovy(h)")

#a_all = avg_temps.find_all('a')[1].text
#td_all = avg_temps.find_all('td')[3].text

datetime='2020-10-23'
data_symbol = {
    "symbol": "",
    "company_name": "", 
    "price": "",
    "day_change": "",
    "percent_change": "",
    "market_cap": "",
    "currency_value": "",
    "currency_value_old": "",
    "total_volume": "",
    "currency_supply": "",
    "trans_date": ""

}
offsetVal=0
for x in range(1, 4):

    
#a_all = avg_temps.find_all('a')[1].text
#td_all = avg_temps.find_all('td')[3].text

    html = browser.html
    soup = bs(html, "html.parser")
    avg_temps = soup.find("div", class_="Ovx(a) Ovx(h)--print Ovy(h)")
    
    vol_str_data = ''
    int_vol_str = ''

    avg_vol_str_data = ''
    avg_int_vol_str = ''

    mkt_vol_str_data = ''
    mkt_int_vol_str = ''

    tot_vol_str_data = ''
    tot_int_vol_str = ''

    sup_vol_str_data = ''
    sup_int_vol_str = ''
    
    paragraphs = avg_temps.find_all('tr')
    for para in paragraphs:
        data_symbol={}
        try:
            data_symbol["symbol"] = para.find_all('td')[0].text
            data_symbol["company_name"] = para.find_all('td')[1].text
            data_symbol["price"] = para.find_all('td')[2].text
            data_symbol["price"] =  data_symbol["price"].replace(',','')
            data_symbol["day_change"] = para.find_all('td')[3].text
            data_symbol["percent_change"] = para.find_all('td')[4].text
            
            vol_str_data= para.find_all('td')[5].text
            vol_str_data=str(vol_str_data).replace(',','')
            if (vol_str_data.find('M') != -1):
                vol_str_data=str(vol_str_data).replace('M','')
                int_vol_str=float(vol_str_data)*1000000
                data_symbol["market_cap"] = int_vol_str
            elif (vol_str_data.find('B') != -1):
                vol_str_data=str(vol_str_data).replace('B','')
                int_vol_str=float(vol_str_data)*1000000000
                data_symbol["market_cap"] = int_vol_str
            elif (vol_str_data.find('T') != -1):
                vol_str_data=str(vol_str_data).replace('T','')
                int_vol_str=float(vol_str_data)*1000000000000
                data_symbol["market_cap"] = int_vol_str
            else:
                data_symbol["market_cap"]=vol_str_data

            avg_vol_str_data= para.find_all('td')[6].text
            avg_vol_str_data=str(avg_vol_str_data).replace(',','')
            if (avg_vol_str_data.find('M') != -1):
                avg_vol_str_data=str(avg_vol_str_data).replace('M','')
                avg_int_vol_str=float(avg_vol_str_data)*1000000
                data_symbol["currency_value"] = avg_int_vol_str
            elif (avg_vol_str_data.find('B') != -1):
                avg_vol_str_data=str(avg_vol_str_data).replace('B','')
                avg_int_vol_str=float(avg_vol_str_data)*1000000000
                data_symbol["currency_value"] = avg_int_vol_str
            elif (avg_vol_str_data.find('T') != -1):
                avg_vol_str_data=str(avg_vol_str_data).replace('T','')
                avg_int_vol_str=float(avg_vol_str_data)*1000000000000
                data_symbol["currency_value"] = avg_int_vol_str
            else:
                data_symbol["currency_value"]=avg_vol_str_data

            mkt_vol_str_data= para.find_all('td')[7].text
            mkt_vol_str_data=str(mkt_vol_str_data).replace(',','')
            if (mkt_vol_str_data.find('B') != -1):
                mkt_vol_str_data=str(mkt_vol_str_data).replace('B','')
                mkt_int_vol_str=float(mkt_vol_str_data)*1000000000
                data_symbol["currency_value_old"] = mkt_int_vol_str
            elif (mkt_vol_str_data.find('M') != -1):
                mkt_vol_str_data=str(mkt_vol_str_data).replace('M','')
                mkt_int_vol_str=float(mkt_vol_str_data)*1000000
                data_symbol["currency_value_old"] = mkt_int_vol_str
            elif (mkt_vol_str_data.find('T') != -1):
                mkt_vol_str_data=str(mkt_vol_str_data).replace('M','')
                mkt_int_vol_str=float(mkt_vol_str_data)*1000000000000
                data_symbol["currency_value_old"] = mkt_int_vol_str
            else:
                data_symbol["currency_value_old"]=mkt_vol_str_data

            tot_vol_str_data= para.find_all('td')[8].text
            tot_vol_str_data=str(tot_vol_str_data).replace(',','')
            if (tot_vol_str_data.find('B') != -1):
                tot_vol_str_data=str(tot_vol_str_data).replace('B','')
                tot_int_vol_str=float(tot_vol_str_data)*1000000000
                data_symbol["total_volume"] = tot_int_vol_str
            elif (tot_vol_str_data.find('M') != -1):
                tot_vol_str_data=str(tot_vol_str_data).replace('M','')
                tot_int_vol_str=float(tot_vol_str_data)*1000000
                data_symbol["total_volume"] = tot_int_vol_str
            elif (tot_vol_str_data.find('T') != -1):
                tot_vol_str_data=str(tot_vol_str_data).replace('M','')
                tot_int_vol_str=float(tot_vol_str_data)*1000000000000
                data_symbol["total_volume"] = tot_int_vol_str
            else:
                data_symbol["total_volume"]=tot_vol_str_data

            sup_vol_str_data= para.find_all('td')[9].text
            sup_vol_str_data=str(sup_vol_str_data).replace(',','')
            if (sup_vol_str_data.find('B') != -1):
                sup_vol_str_data=str(sup_vol_str_data).replace('B','')
                sup_int_vol_str=float(sup_vol_str_data)*1000000000
                data_symbol["currency_supply"] = sup_int_vol_str
            elif (sup_vol_str_data.find('M') != -1):
                sup_vol_str_data=str(sup_vol_str_data).replace('M','')
                sup_int_vol_str=float(sup_vol_str_data)*1000000
                data_symbol["currency_supply"] = sup_int_vol_str
            elif (sup_vol_str_data.find('T') != -1):
                sup_vol_str_data=str(sup_vol_str_data).replace('M','')
                sup_int_vol_str=float(sup_vol_str_data)*1000000000000
                data_symbol["currency_supply"] = sup_int_vol_str
            else:
                data_symbol["currency_supply"]=sup_vol_str_data
                
                
            data_symbol["trans_date"]=datetime

            #     vol_str=vol_str.replace('M','')
            #    vol_str.replace('M','')
                #int_vol_str=vol_str.to_numeric()
                #int_vol_str=int_vol_str*1000000
            #data_symbol["volume"] = para.find_all('td')[5].text
            #data_symbol["avg_volume"] = para.find_all('td')[6].text
            #data_symbol["market_cap"] = para.find_all('td')[7].text
            data_array.append(data_symbol)
        except:
            pass
    #browser.find_by_text('Next').first
    offsetVal=offsetVal+25
    url = "https://finance.yahoo.com/cryptocurrencies?count=25&offset="+str(offsetVal)
    browser.visit(url)
    time.sleep(1)
#browser.links.find('div', class_='O(n):f O(n):h P(0) M(0) Cur(p):h D(ib)')
#find_element_by_link_text
#find_by_name('name').last
#find_by_partial_text
#vol_str = vol_str.astype('string')
#int_data=''
#s.find("is") == -1:
#if (vol_str_data.find('M') != -1):
#    vol_str_data = vol_str_data.replace('M','')
#    val_data=float(vol_str_data)*1000000
data_array
```

```python
data_array_df = pd.DataFrame(data_array)
data_array_df
```
