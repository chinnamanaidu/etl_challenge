### This is notebook to process yahoo most active transaction symbols

```python
import pandas as pd
from sqlalchemy import create_engine
from splinter import Browser
from bs4 import BeautifulSoup as bs
import time
```

##### Store CSV into DataFrame

```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

##### Initialize the browser objects

```python
# Visit visitcostarica.herokuapp.com
url = "https://finance.yahoo.com/most-active"
browser.visit(url)

time.sleep(1)

# Scrape page into Soup
#html = browser.html
#soup = bs(html, "html.parser")
```

#### This is the iteration process to screen scrap the data from yahoo website, there is two level iteration first level is for each page and second level on each page to get the segment of data

```python
data_array = []
#avg_temps = soup.find("div", class_="Ovx(a) Ovx(h)--print Ovy(h)")

#a_all = avg_temps.find_all('a')[1].text
#td_all = avg_temps.find_all('td')[3].text

datetime='2020-10-23'
data_symbol = {
    "symbol": "",
    "company_name": "", 
    "price": "",
    "day_change": "",
    "percent_change": "",
    "volume": "",
    "avg_volume": "",
    "market_cap": "",
    "trans_date": ""
}

offsetVal=0
for x in range(1, 6):

    html = browser.html
    soup = bs(html, 'html.parser')



    avg_temps = soup.find("div", class_="Ovx(a) Ovx(h)--print Ovy(h)")

#a_all = avg_temps.find_all('a')[1].text
#td_all = avg_temps.find_all('td')[3].text
    vol_str_data = ''
    int_vol_str = ''

    avg_vol_str_data = ''
    avg_int_vol_str = ''

    mkt_vol_str_data = ''
    mkt_int_vol_str = ''

    paragraphs = avg_temps.find_all('tr')
    for para in paragraphs:
        data_symbol={}
        try:
            data_symbol["symbol"] = para.find_all('td')[0].text
            data_symbol["company_name"] = para.find_all('td')[1].text
            data_symbol["price"] = para.find_all('td')[2].text
            data_symbol["day_change"] = para.find_all('td')[3].text
            data_symbol["percent_change"] = para.find_all('td')[4].text
            vol_str_data= para.find_all('td')[5].text
            if (vol_str_data.find('M') != -1):
                vol_str_data=str(vol_str_data).replace('M','')
                int_vol_str=float(vol_str_data)*1000000
                data_symbol["volume"] = int_vol_str
            else:
                data_symbol["volume"]=para.find_all('td')[5].text

            avg_vol_str_data= para.find_all('td')[6].text
            if (avg_vol_str_data.find('M') != -1):
                avg_vol_str_data=str(avg_vol_str_data).replace('M','')
                avg_int_vol_str=float(avg_vol_str_data)*1000000
                data_symbol["avg_volume"] = avg_int_vol_str
            else:
                data_symbol["avg_volume"]=para.find_all('td')[6].text

            mkt_vol_str_data= para.find_all('td')[7].text
            if (mkt_vol_str_data.find('B') != -1):
                mkt_vol_str_data=str(mkt_vol_str_data).replace('B','')
                mkt_int_vol_str=float(mkt_vol_str_data)*1000000000
                data_symbol["market_cap"] = mkt_int_vol_str
            elif (mkt_vol_str_data.find('M') != -1):
                mkt_vol_str_data=str(mkt_vol_str_data).replace('M','')
                mkt_int_vol_str=float(mkt_vol_str_data)*1000000
                data_symbol["market_cap"] = mkt_int_vol_str
            elif (mkt_vol_str_data.find('T') != -1):
                mkt_vol_str_data=str(mkt_vol_str_data).replace('M','')
                mkt_int_vol_str=float(mkt_vol_str_data)*1000000000000
                data_symbol["market_cap"] = mkt_int_vol_str
            else:
                data_symbol["market_cap"]=para.find_all('td')[7].text

            data_symbol["trans_date"]=datetime

            #     vol_str=vol_str.replace('M','')
            #    vol_str.replace('M','')
                #int_vol_str=vol_str.to_numeric()
                #int_vol_str=int_vol_str*1000000
            #data_symbol["volume"] = para.find_all('td')[5].text
            #data_symbol["avg_volume"] = para.find_all('td')[6].text
            #data_symbol["market_cap"] = para.find_all('td')[7].text
            data_array.append(data_symbol)
        except:
            pass
    #browser.find_by_text('Next').first
    offsetVal=offsetVal+25
    url = "https://finance.yahoo.com/most-active?count=25&offset="+str(offsetVal)
    browser.visit(url)
    time.sleep(2)
#browser.links.find('div', class_='O(n):f O(n):h P(0) M(0) Cur(p):h D(ib)')
#find_element_by_link_text
#find_by_name('name').last
#find_by_partial_text
#vol_str = vol_str.astype('string')
#int_data=''
#s.find("is") == -1:
#if (vol_str_data.find('M') != -1):
#    vol_str_data = vol_str_data.replace('M','')
#    val_data=float(vol_str_data)*1000000
data_array
```

##### Change the data to dataframe

```python
data_array_df = pd.DataFrame(data_array)
data_array_df
```

```python
rds_connection_string = "postgres:admin@localhost:5432/finance_db"
#<insert password>@localhost:5432/customer_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```
